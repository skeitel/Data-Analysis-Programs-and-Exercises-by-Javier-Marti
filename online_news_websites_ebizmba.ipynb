```python
df.dropna(how = 'any', inplace = True)
df
```

```python
#change all cols to numeric
col_list = ['Compete rank', 'M Unique visitors', 'Quantcast rank']
for item in col_list:
    df[item] = pd.to_numeric(df[item], errors='coerce')
```

```python
df.describe()
```

```python
fig,ax = plt.subplots(figsize=(16,9))
plt.bar(df['Company'], df['M Unique visitors'], color = 'cyan', label = 'Visitors')
plt.grid()
ax.legend()
plt.title('Unique monthly visitors')
```

```python
fig,ax = plt.subplots(figsize=(16,9))
plt.plot(df['Company'], df['Compete rank'], color = 'green', label = 'Compete rank (lower better)')
plt.plot(df['Company'], df['Quantcast rank'], color = 'red', label = 'Quantcast rank (higher better)')
plt.grid()
ax.legend()
plt.title('Comcast and Compete rank')

#Get second axis
ax2 = ax.twinx()
plt.plot(df['Company'], df['M Unique visitors'], color = 'black', label = 'Unique M visitors')
ax2.legend(loc='best')
ax2.tick_params('vals', colors = 'b')
```

```python
#Finding the BEST BALANCE between lowest Compete rank and highest Quantcast rank
df_balance = df.groupby('Company').agg({'Compete rank': 'min', 'Quantcast rank':'max'})[['Compete rank', 'Quantcast rank']].reset_index()
df_balance
```

```python
#Creating best rank ration column to calculate the same as above in a different way
df['best rank ratio'] = df['Quantcast rank'] - df['Compete rank']
```

```python
#Showing best rank ratio
df.sort_values('best rank ratio', ascending = False).head()
```

```python
fig,ax = plt.subplots(figsize=(16,9))
plt.plot(df.Company, df['best rank ratio'], color = 'red', label = 'Best rank ratio (higher better)')
plt.grid()
ax.legend()
plt.title('Best rank ratio')

#Get second axis
ax2 = ax.twinx()
plt.plot(df['Company'], df['M Unique visitors'], color = 'black', label = 'Unique M visitors')
ax2.legend(loc='best', )
ax2.tick_params('vals', colors = 'b')
```
