```python
#Plot the TOP 25 by TWFolls
df_top25 = df.sort_values(by='TWFolls', ascending = False)
df_top25 = df_top25.iloc[:25, :]

fig,ax = plt.subplots(figsize = (16,9))
plt.plot(df.Site[:25], df_top25.TWFolls, color = 'red', label = 'TWFolls')
plt.bar(df.Site[:25], df_top25.FBFolls, color = 'blue', label = 'FBFolls')

#Rotate xticks
plt.xticks(rotation = 90)
plt.rcParams['xtick.labelsize'] = 10
```

```python
#Finding the BEST BALANCE between max followers on TW and FB
df_balance = df.groupby('Site').agg({'FBFolls': 'max', 'TWFolls':'max'})[['FBFolls', 'TWFolls']].reset_index()
df_balance.head()
```
